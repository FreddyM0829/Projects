```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('meta_prices.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
np.set_printoptions(precision=2)
```

```python
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .08, random_state = 0)
print(X_test)
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
z_pred = regressor.predict(X_test)
print(np.concatenate((z_pred.reshape(len(z_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X_train)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y_train)
```

Plot Result

```python
a = [1,2,3,4,5,6,7,8]
plt.scatter(a, y_test, color = 'red')
plt.scatter(a, z_pred, color = 'blue')
plt.xlabel('Data Points')
plt.ylabel('Cost')
plt.show()
```

Polynomial Regression result

```python
plt.scatter(X_train, y_train, color = 'red')
plt.scatter(X_train, lin_reg_2.predict(poly_reg.fit_transform(X_train)), color = 'blue')
plt.title('Polynomial Regression')
plt.xlabel('High cost')
plt.ylabel('Low cost')
plt.show()
```

```python

plt.scatter(X_test, y_test, color = 'red')
plt.scatter(X_test, lin_reg_2.predict(poly_reg.fit_transform(X_test)), color = 'blue')
plt.title('Polynomial Regression')
plt.xlabel('High cost')
plt.ylabel('Low cost')
plt.show()
```
